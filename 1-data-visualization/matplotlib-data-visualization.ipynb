# Data visualization with Matplotlib

```python
import matplotlib.pyplot as plt
```

# Line Plot

```python
x_days = [1, 2, 3, 4, 5]
```

```python
y_price1 = [9, 9.5, 10.1, 10, 12]
y_price2 = [11, 12, 10.5, 11.5, 12.5]
```

```python
plt.title('Stock Movement')
plt.xlabel('Week Days')
plt.ylabel('Price in USD')

plt.plot(x_days, y_price1, label='Stock 1')
plt.plot(x_days, y_price2, label='Stock 2')

plt.legend(loc=2, fontsize=12)

plt.show()
```

# Bar Chart

```python
# Create a list of cities and temperatures
x_cities = ['New York', 'London', 'Dubai', 'Paris', 'Tokyo']
y_temp = [75, 65, 105, 63, 90]
```

```python
# Chart labels
plt.title('Temperature Variations')
plt.xlabel('Cities')
plt.ylabel('Temperature')

# Create the Bar chart
plt.bar(x_cities, y_temp)

# Show the plot
plt.show()
```

# Histogram

```python
# Open and read the file
f = open('agedata.csv','r')
agefile = f.readlines()
```

# Pie Chart

```python
# Open the file and read the data
f = open('agedata2.csv','r')
agefile = f.readlines()
```

```python
# Load the records to a list and split them to create list of cities
city_list = []

for data in agefile:
    age, city = data.split(sep=',')
    city_list.append(city)
    
# Create a dictionary of City Names and Total count per city
from collections import Counter
city_count = Counter(city_list)
city_count
```

```python
# Unpack city names and total count to respective lists
city_names = list(city_count.keys())
city_values = list(city_count.values())

print(city_names)
print(city_values)
```

```python
#Create the pie chart
plt.pie(city_values, labels=city_names, autopct='%.2f%%')
plt.show()
```

# Scatter Plot

```python
# Open the file in read mode and read lines
f = open('salesdata2.csv','r')
salefile = f.readlines()

salefile
```

```python
# Create the sales List
s_list = []
c_list = []

# Append all the records from the file to the saleslist
for data in salefile:
    sale, cost = data.split(sep=',')
    s_list.append(int(sale))
    c_list.append(int(cost))
    
print(s_list)
print(c_list)
```

```python
# Change the chart labels
plt.title("Sales Vs Cost")
plt.xlabel("Sale")
plt.ylabel("Cost")

# Create the scatter plot
plt.scatter(s_list,c_list)
plt.show()
```

# Multiple plots in the same figure

```python
# FIGURE 1 **** Box Plot
# Open the file in read mode and read lines
f = open('salesdata.csv','r')
salefile = f.readlines()
sale_list = [int(sale) for sale in salefile]

plt.subplot(2, 2, 1) # this is the 1st plot of our 2x2 figure
plt.title("Box Plot of Sales")
plt.boxplot(sale_list)




# FIGURE 2 **** Line chart
x_days = [1, 2, 3, 4, 5]

y_price1 = [9, 9.5, 10.1, 10, 12]
y_price2 = [11, 12, 10.5, 11.5, 12.5]

plt.subplot(2, 2, 2) # this is the 2nd plot of our 2x2 figure
plt.title('Stock Movement')
plt.xlabel('Week Days')
plt.ylabel('Price in USD')

plt.plot(x_days, y_price1, label='Stock 1')
plt.plot(x_days, y_price2, label='Stock 2')

plt.legend(loc=2, fontsize=12)



# FIGURE 3 **** Bar chart
x_cities = ['New York', 'London', 'Dubai', 'Paris', 'Tokyo']
y_temp = [75, 65, 105, 63, 90]

plt.subplot(2, 2, 3) # this is the 3rd plot of our 2x2 figure
plt.title('Temperature Variations')
plt.xlabel('Cities')
plt.ylabel('Temperature')

plt.bar(x_cities, y_temp)



# FIGURE 4 **** Histogram chart
f = open('agedata.csv','r')
agefile = f.readlines()

age_list = [int(age) for age in agefile]
bins = [x for x in range(0,110,10)]

plt.subplot(2, 2, 4) # this is the 4th plot of our 2x2 figure
plt.title("Age histogram")
plt.xlabel("Group")
plt.ylabel("Age")

plt.hist(age_list, bins, histtype='bar', rwidth=0.9)

# Addressing the overlapping plot layout issue
plt.tight_layout()

# Save the figure in one image
plt.savefig('multi-plot-figure.png')

# Show the Plot
plt.show()
```
