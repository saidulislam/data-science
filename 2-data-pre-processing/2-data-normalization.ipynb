```python
import pandas as pd
```

```python
# read CSV file using pandas
df = pd.read_csv('loan_small.csv')
df
```

```python
# drop the rows with the missing values
cleandata = df.dropna()
cleandata
```

```python
# operate on the 3 numerical columns
num_data = cleandata.iloc[:, 2:5]
num_data
```

# Normalize the data with zscaler method

```python
# import standard scaler from sklearn
from sklearn.preprocessing import StandardScaler

# this will normalize the data with zscaler method
scaler_obj = StandardScaler()
scalar_data = scaler_obj.fit_transform(num_data)
scalar_data
```

# Normalize the data with minmax method

```python
# import standard scaler from sklearn
from sklearn.preprocessing import minmax_scale

mm_scaler_data = minmax_scale(num_data) # default 0-1 for minmax. you can provide your own range
mm_scaler_data
```
